```python
#number of unique users in the dataset
len(books.uid.unique())
```

```python
len(books.Title.unique())
```

```python
user_book_df = books.pivot_table(index='uid',
                                 columns='Title',
                                 values='Rating',aggfunc=len)
```

```python
user_book_df
```

```python
user_book_df.index = books.uid.unique()
```

```python
user_book_df
```

```python
#Impute those NaNs with 0 values
user_book_df.fillna(0, inplace=True)
```

```python
user_book_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_book_df.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = books.uid.unique()
user_sim_df.columns =books.uid.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:10, 0:10]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:10]
```
